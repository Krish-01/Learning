### Creating a dataframe of the columns which are of Object dataype:

#### Getting the columns which are of Object datatype/categorical data:

```python
df.dtypes=='object' #return True for the columns which is of Object type and for other columns it gives False.
```

```python
df.dtypes[df.dtypes=='object'] #We need the dataype of the columns which are object type
```

**Note:** In the above code, when we pass "df.dtypes == 'object'" to df.dtypes, it will automatically filter  "df.dtypes == 'object'" for True and returns the related output.
>Which means it returns the data of the columns for which  "df.dtypes == 'object'" is True

```python
df.dtypes[df.dtypes=='object'].index # columns which are of Object datatype
```

```python
df[df.dtypes[df.dtypes=='object'].index] #creating a dataframe of the columns which are of Object dataype
```

**Note:**
> df.dtypes[df.dtypes == 'object'].index --- This gives the columnnames.

>df[df.dtypes[df.dtypes == 'object'].index] is equivalent to df['columnname']
so we are getting the data of those columns only.

```python
df1 = df[df.dtypes[df.dtypes=='object'].index]
```

    top gives the highest counted value of the categorical values.

4. frequency - it gives the number of times the top data is present in the dataset.
